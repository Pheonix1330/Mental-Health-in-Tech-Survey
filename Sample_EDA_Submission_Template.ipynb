# **Project Summary -**



This project focuses on performing Exploratory Data Analysis (EDA) on a mental health survey dataset collected from individuals working in the tech industry. The main objective of the project is to analyze and understand the factors that influence mental health conditions and treatment-seeking behavior among employees. Mental health has become an important concern in modern workplaces, especially in high-pressure industries like technology, making this analysis highly relevant for organizations.

The dataset includes a variety of features such as demographic information (age, gender, country), workplace-related factors (company size, remote work, benefits), and mental health-related attributes (family history, treatment, work interference, and awareness programs). These variables provide a comprehensive view of both personal and organizational aspects affecting mental health.

The first step in the project involved data cleaning and preprocessing to make the dataset suitable for analysis. Missing values were handled appropriately by replacing them with meaningful placeholders or removing irrelevant entries. Duplicate records were identified and removed to ensure data accuracy. Inconsistent values, especially in categorical variables like gender, were standardized to maintain uniformity. Additionally, unrealistic values such as incorrect age entries were filtered out to improve data quality.

After preparing the dataset, various exploratory techniques were applied to extract meaningful insights. Univariate analysis was conducted to understand the distribution of individual variables, such as age, gender, and treatment status. Bivariate analysis was then used to study relationships between two variables, such as gender vs treatment and family history vs treatment. Multivariate analysis further explored complex relationships involving multiple variables, such as age, gender, and treatment combined.

Through these analyses, several important insights were identified. It was observed that most respondents belong to the younger working-age group, indicating that mental health concerns are more prominent among early and mid-career professionals. The dataset also showed a gender imbalance, with a higher number of male respondents. One of the key findings was that individuals with a family history of mental illness are more likely to seek treatment. Additionally, many respondents reported that mental health issues interfere with their work, highlighting its direct impact on productivity.

Another significant observation was the role of workplace support systems. Employees who have access to mental health benefits and resources are more likely to seek help. However, lack of awareness about available support options reduces the effectiveness of these benefits. Remote work also showed some influence on mental health, suggesting the need for proper work-life balance and support mechanisms in flexible work environments.

The insights derived from this project can help organizations make informed decisions to improve employee well-being. By promoting mental health awareness, providing accessible support systems, and encouraging open communication, companies can create a healthier work environment. This not only benefits employees but also leads to increased productivity, reduced absenteeism, and better overall organizational performance.

In conclusion, this project demonstrates the importance of data-driven analysis in understanding mental health trends in the workplace. The findings emphasize the need for proactive measures and supportive policies to address mental health challenges effectively.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# ==============================
# Data Cleaning & Preparation
# ==============================

# Create a copy of dataset to avoid modifying original data
df_clean = df.copy()   # Keeps original dataset safe

# ------------------------------
# Handle Missing Values
# ------------------------------

# Fill missing values in categorical columns with 'Unknown'
for col in df_clean.select_dtypes(include='object').columns:
    df_clean[col] = df_clean[col].fillna('Unknown')   # Replace NaN with 'Unknown'

# ------------------------------
# Clean Age Column
# ------------------------------

# Remove unrealistic age values
df_clean = df_clean[(df_clean['Age'] >= 18) & (df_clean['Age'] <= 60)]   # Keep valid age range

# ------------------------------
# Standardize Gender Column
# ------------------------------

# Convert all gender values to lowercase for consistency
df_clean['Gender'] = df_clean['Gender'].str.lower()

# Replace common variations with standard labels
df_clean['Gender'] = df_clean['Gender'].replace({
    'male': 'Male', 'm': 'Male', 'man': 'Male',
    'female': 'Female', 'f': 'Female', 'woman': 'Female'
})

# Replace remaining values as 'Other'
df_clean['Gender'] = df_clean['Gender'].apply(lambda x: x if x in ['Male', 'Female'] else 'Other')

# ------------------------------
# Remove Duplicate Rows
# ------------------------------

df_clean = df_clean.drop_duplicates()   # Remove duplicate records

# ------------------------------
# Final Check
# ------------------------------

# Display cleaned dataset shape
print("Cleaned Dataset Shape:", df_clean.shape)

# Preview cleaned data
df_clean.head()
````

### What all manipulations have you done and insights you found?

Data Manipulations:

Handled missing values by replacing them with appropriate values (e.g., "Unknown")
Removed duplicate records to ensure data accuracy
Cleaned inconsistent entries in categorical variables (e.g., Gender standardization)
Filtered unrealistic values (e.g., invalid age entries)
Converted data into a consistent and analysis-ready format

Insights Found:

Majority of respondents belong to the working-age group
Gender distribution is imbalanced

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
# Create a copy for encoding
df_encoded = df_clean.copy()   # Copy dataset

# Convert categorical columns to numeric
for col in df_encoded.select_dtypes(include='object').columns:
    df_encoded[col] = df_encoded[col].astype('category').cat.codes   # Encode categories

# Create correlation matrix
corr = df_encoded.corr()   # Compute correlations

# Plot heatmap
plt.figure(figsize=(12,8))   # Set figure size

sns.heatmap(corr,            # Correlation data
            cmap='coolwarm', # Color map
            annot=False)     # Disable numbers for clarity

# Add title
plt.title("Correlation Heatmap")

# Show plot
plt.show()
````

##### 1. Why did you pick the specific chart?

Heatmap helps visualize relationships between multiple variables at once.

##### 2. What is/are the insight(s) found from the chart?

Some variables show moderate correlation with treatment (like family history)

Most features have weak correlation, indicating complex relationships
